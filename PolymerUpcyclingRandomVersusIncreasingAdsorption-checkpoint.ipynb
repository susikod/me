Numerical chopping models with different weighting schemes to compare an equal probability of any molecule of linear polyethylene in the melt of being hydrogenolyzed by the catalyst ('numrandc9') to where the probability scales linearly with the molecular weight of the alkane ('wtrandnc9'). In both models, the location of the chop in the alkane is random, as it is assumed that all points on the alkane molecule are equally likely to interact with the catalytic site. Molecules with fewer than nine carbons are removed from the model, as they are assumed to be in the gas phase and less likely to interact with the catalyst than alkanes in the melt. The model 'numrandall' is included to show completely random chopping without removing alkanes below C9.

```python
# using Python 3.7
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
# load in PE distribution generated from GPC data
# the PE distribution is proportional to the GPC data from the PE 
# purchased from Sigma-Aldrich (Product number = 427799)
# using the same integration window in the attached PDF of GPC data
file = 'InitialPE7700_CarbonNumber_v3.csv'
PEhist = np.genfromtxt(file,delimiter=',',dtype=int,encoding='utf-8-sig')
PElinear = []
for i in range(len(PEhist)):
    for j in range(PEhist[i,1]):
        PElinear.append(PEhist[i,0])
PElinear = np.array(PElinear)
```

```python
def plotbar(PE,title,MN,MW,PDI,yessave = False,file_save = 'testbar'):
    unique, counts = np.unique(PE, return_counts=True)
    
    wide = np.true_divide(unique,10)
    fig1 = plt.figure()
    ax = fig1.add_subplot(111)
    ax.bar(unique,counts,width=wide)
    
    # log scale to match standard GPC plotting
    ax.set_xscale('log')
    ax.set_xlim((100,1000000))
    
    ax.text(0.8,0.9,'MN: '+str(round(MN,1)),
             horizontalalignment='center',
             verticalalignment='center',
             transform=ax.transAxes)
    ax.text(0.8,0.8,'MW: '+str(round(MW,1)),
             horizontalalignment='center',
             verticalalignment='center',
             transform=ax.transAxes)
    ax.text(0.8,0.7,'PDI: '+str(round(PDI,4)),
             horizontalalignment='center',
             verticalalignment='center',
             transform=ax.transAxes)
    ax.set_title(title)
    plt.show()
    
    if yessave == True:   
        fig1.savefig(file_save,bbox_inches='tight',dpi=300,transparent=True)
```

```python
def plotdistro(distro,yessave = False,file_save = 'testdistro'):
    #print plot of distribution versus time
    #print(distro[:,0])
    fig1 = plt.figure()
    ax1 =fig1.add_subplot(121)
    ax1.plot(distro[:,0],distro[:,1:3])
    ax1.set_xlabel('Chop (#)')
    ax1.set_ylabel('MW (Da)')
    
    ax1.tick_params(axis='x',labelrotation=45)

    ax2 = ax1.twinx()
    ax2.plot(distro[:,0],distro[:,3],color='r')
    ax2.set_ylabel('PDI')
    ax2.set_ylim(bottom=0)

    ax3 = fig1.add_subplot(122)
    ax3.plot(distro[:,0],distro[:,4],color='g')
    ax3.set_xlabel('Chop (#)')
    ax3.set_ylabel('Molecules (n)')
    
    ax3.tick_params(axis='x',labelrotation=45)
    
    plt.tight_layout()
    
    if yessave == True:   
        fig1.savefig(file_save,bbox_inches='tight',dpi=300,transparent=True)
```

```python
#PE sample in units of carbons
cPE = np.copy(PElinear)
#convert from carbon# to MW
PE = np.add(np.multiply(cPE,14),2)
PE = PE.astype('int64')

# calculate molecular weights and PDI
Mn = len(PE)/np.sum(np.true_divide(1,PE))
Mw = np.sum(PE)/len(PE)
PDI = Mw/Mn
#print(np.sum(PE))
#print(np.true_divide(1,PE))
#print(np.power(PE,2))
#print(PE.dtype)
print(Mn,Mw,PDI)

```

```python
# check that PE distribution has loaded correctly
filename = 'initial'
plotbar(PE,'InitialPE',Mn,Mw,PDI,yessave=True,file_save=filename)
```

```python
# random chop
# include gas phase products
fhead = 'numrandall'

# number of runs (best to keep as multiple of 10)
runnumb = 80000

# cycle counter for frequency to sample MN, MW, PDI
nowcalc = 500

# cycle counter for frequency to plot bars
nowplot = 5000

#save bars
plotsave = True

#copy over in units of carbon
chopcPE = np.copy(cPE)

#convert from carbons to Da
chopPE = np.add(np.multiply(chopcPE,14),2)

chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
chopMw = np.sum(chopPE)/len(chopPE)
chopPDI = chopMw/chopMn
chopN = len(chopPE)

distro = np.array([[0, chopMn, chopMw, chopPDI, chopN]])


#print initial bar
'''
plotbar(chopPE,'Initial',chopMn,chopMw,chopPDI)

'''

for n in range(runnumb):
    #find PE strand to chop
    randspot = np.random.randint(0,len(chopcPE))
    #print(randspot,chopcPE[randspot])
    
    #chop chosen strand
    if chopcPE[randspot] ==1:
        chopA = 1
        chopB = False
    
    else:
        chopA = np.random.randint(1,chopcPE[randspot])
        chopB = chopcPE[randspot]-chopA
    #print(chopA,chopB)   
    
    #remove chopped strand
    chopcPE = np.delete(chopcPE,randspot)
        
    #add new strands
    if chopA ==1 and chopB ==False:
        chopcPE = np.append(chopcPE,[chopA])
    else:
        chopcPE = np.append(chopcPE,[chopA,chopB])
    #print(len(chopcPE))
    
    #convert from carbon# to MW
    chopPE = np.add(np.multiply(chopcPE,14),2)
    
    #use modulus to set print rate
    
    if n%nowcalc ==0:
        #calc new Mn and Mw
        chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
        chopMw = np.sum(chopPE)/len(chopPE)
        chopPDI = chopMw/chopMn
        chopN = len(chopPE)
        #add new product distributions to array
        distro = np.append(distro,[[n+1, chopMn, chopMw, chopPDI, chopN]], axis=0)
        
        #print(chopMn,chopMw,chopPDI)
        
        if n%nowplot ==0:
            #plot bar
            plotbar(chopPE,'Chop '+str(n+1),chopMn,chopMw,chopPDI,
                     yessave=plotsave, file_save=fhead+'chop'+str(n+1))
    

# print end state MW    
chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
chopMw = np.sum(chopPE)/len(chopPE)
chopPDI = chopMw/chopMn
chopN = len(chopPE)
#add new product distributions to array
distro = np.append(distro,[[runnumb, chopMn, chopMw, chopPDI, chopN]],axis=0)
#print(chopMn,chopMw,chopPDI)

#print end state bar
plotbar(chopPE,'Chop '+str(runnumb),chopMn,chopMw,chopPDI,
         yessave=plotsave, file_save=fhead+'chop'+str(runnumb))


#print change in Mn Mw and PDI
plotdistro(distro, yessave=plotsave, file_save=fhead+'distro')    
```

```python
# weighted random chop
# weigh the probablility to chop each strand by the number of carbons in it
# weight such that weighing is proportional to n
# exclude products under C9 as entering gas phase and not contributing to product measured by GPC
fhead = 'wtrandnc9'


# number of runs (best to keep as multiple of 10)
runnumb = 80000

# cycle counter for frequency to sample MN, MW, PDI
nowcalc = 500

# cycle counter for frequency to plot bars
nowplot = 5000

#save bars
plotsave = True

# load in initial PE state in carbon#
chopcPE = np.copy(cPE)

#convert from carbons to Da
chopPE = np.add(np.multiply(chopcPE,14),2)

chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
chopMw = np.sum(chopPE)/len(chopPE)
chopPDI = chopMw/chopMn
chopN = len(chopPE)

distro = np.array([[0, chopMn, chopMw, chopPDI, chopN]])


#print initial bar
'''
plotbar(chopPE,'Initial',chopMn,chopMw,chopPDI)

'''

for n in range(runnumb):
    #Weighing term for probability of choosing a polymer - equal to the number of carbons
    weight_prob = np.copy(chopcPE/np.sum(chopcPE))
    
    #find PE strand to chop
    randspot = np.random.choice(len(chopcPE),p=weight_prob)
    #print(randspot,chopcPE[randspot])
    
    #chop chosen strand
    if chopcPE[randspot] ==1:
        chopA = 1
        chopB = False
    
    else:
        chopA = np.random.randint(1,chopcPE[randspot])
        chopB = chopcPE[randspot]-chopA
    #print(chopA,chopB)   
    
    #remove chopped strand
    chopcPE = np.delete(chopcPE,randspot)
        
    #add new strands
    #exclude below C9
    if chopA ==1 and chopB ==False:
        if chopA > 8:
            chopcPE = np.append(chopcPE,[chopA])
    else:
        if chopA > 8 and chopB > 8:
            chopcPE = np.append(chopcPE,[chopA,chopB])
        elif chopA >8:
            chopcPE = np.append(chopcPE,[chopA])
        elif chopB >8:
            chopcPE = np.append(chopcPE,[chopB])
    #print(len(chopcPE))
    
    #convert from carbon# to MW
    chopPE = np.add(np.multiply(chopcPE,14),2)
    
    #use modulus to set plot rate
    if n%nowcalc == 0: 
        #calc new Mn and Mw
        chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
        chopMw = np.sum(chopPE)/len(chopPE)
        chopPDI = chopMw/chopMn
        chopN = len(chopPE)
        #add new product distributions to array
        distro = np.append(distro,[[n+1, chopMn, chopMw, chopPDI, chopN]], axis=0)
        #print(chopMn,chopMw,chopPDI)
        
        
        if n%nowplot ==0:
            #plot bar
            plotbar(chopPE,'Chop '+str(n+1),chopMn,chopMw,chopPDI,
                     yessave=plotsave, file_save=fhead+'chop'+str(n+1))        

#print number in initial
print(len(cPE))
#print number in final
print(len(chopcPE))
#print smallest carbon chain length
print(np.amin(chopcPE))
#print number with that chain length
print((chopcPE == np.amin(chopcPE)).sum())
    
# print end state MW    
chopMn = len(chopPE)/np.sum(np.true_divide(1,chopPE))
chopMw = np.sum(chopPE)/len(chopPE)
chopPDI = chopMw/chopMn
chopN = len(chopPE)
#add new product distributions to array
distro = np.append(distro,[[runnumb, chopMn, chopMw, chopPDI, chopN]],axis=0)
#print(chopMn,chopMw,chopPDI)    
        
#print end state bar
plotbar(chopPE,'Chop '+str(runnumb),chopMn,chopMw,chopPDI,
         yessave=plotsave, file_save=fhead+'chop'+str(runnumb))
 
#print change in Mn Mw and PDI
plotdistro(distro, yessave=plotsave, file_save=fhead+'distro')    
```
